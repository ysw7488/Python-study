```python
import matplotlib.pyplot as plt
```

```python
year, number = (['1st Year', '2nd Year', '3nd Year', '4th Year'], [14, 11, 15, 18])

plt.figure(figsize=(12, 3))

plt.bar(year, number)

plt.show()
```

```python
plt.figure(figsize=(12, 3))

plt.subplot(1, 2, 1)

plt.bar(year, number)

plt.subplot(1, 2, 2)

plt.barh(year, number)

plt.show()
```
